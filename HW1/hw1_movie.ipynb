```python
import pandas as pd
df = pd.read_csv('IMDB-Movie-Data.csv')
result1 = df[df["Year"]==2016].sort_values(by = "Rating",ascending=False).head(3)
print("Top‐3 movies with the highest ratings in 2016",end="?\n")
result1.loc[:,["Title"]]
```

```python
Actor = df["Actors"].str.split("|")
ActorRevenue = []
for i in range(0,2):
    ActorRevenue.append([])
for i in range(len(Actor)):#把每個演員賺了多少錢放入陣列
    for j in range(len(Actor[i])):
        Actor[i][j] = Actor[i][j].strip()
        ActorRevenue[0].append(Actor[i][j])
        ActorRevenue[1].append(float(df["Revenue (Millions)"].iloc[i]))
ActorRevenue=pd.DataFrame(ActorRevenue,index=['Actor','Revenue'])#建立演員對收入的dataframe
ActorRevenue = ActorRevenue.T
ActorRevenue['Revenue'] = ActorRevenue.Revenue.astype('float')
gp = ActorRevenue.groupby(by='Actor')
ActorRevenue['AvgRevenue'] = gp['Revenue'].transform('mean')
ActorRevenue = ActorRevenue.drop("Revenue",axis=1).drop_duplicates()
print("The actor generating the highest average revenue",end="?\n")
ActorRevenue.nlargest(columns='AvgRevenue',n=1,keep='all')
```

```python
TotalRating=0
k=0#計數器
for i in range(0,len(df.index)):#計算艾瑪華森出演電影之評價加總
    if df['Actors'].iloc[i].find("Emma Watson")!=-1:
        TotalRating = TotalRating + float(df['Rating'].iloc[i])
        k = k+1
AvgRating = TotalRating/k  #計算平均
print("The average rating of Emma Watson’s movies",end="?\n")
print(AvgRating)
```

```python
DirectorActor = []
for i in range(0,2):
    DirectorActor.append([])
for i in range(len(Actor)):
    for j in range(len(Actor[i])):
        Actor[i][j] = Actor[i][j].strip()
        DirectorActor[0].append(Actor[i][j])
        DirectorActor[1].append(df["Director"].iloc[i])
DirectorActor=pd.DataFrame(DirectorActor,index=['Actor','Director'])
DirectorActor = DirectorActor.T.drop_duplicates()#移除重複項
gp = DirectorActor.groupby(by='Director')
print("Top-3 directors who collaborate with the most actors",end="?\n")
gp.count().nlargest(columns= 'Actor',n=3,keep='all')
```

```python
ActorGenre = []
Genre = df["Genre"].str.split("|")
for i in range(0,2):
    ActorGenre.append([])
for i in range(len(Actor)):
    for j in range(len(Actor[i])):
        Actor[i][j] = Actor[i][j].strip()
        for k in range(len(Genre[i])):
            ActorGenre[0].append(Actor[i][j])
            ActorGenre[1].append(Genre[i][k])
ActorGenre=pd.DataFrame(ActorGenre,index=['Actor','Genre'])
ActorGenre = ActorGenre.T.drop_duplicates()#移除重複項
gp = ActorGenre.groupby(by='Actor')
print("Top-2 actors playing in the most genres of movies",end="?\n")
gp.count().nlargest(columns='Genre',n=2,keep='all')
```

```python
ActorYear = []
for i in range(0,2):
    ActorYear.append([])
for i in range(len(Actor)):
    for j in range(len(Actor[i])):
        Actor[i][j] = Actor[i][j].strip()
        ActorYear[0].append(Actor[i][j])
        ActorYear[1].append(df["Year"].iloc[i])
ActorYear=pd.DataFrame(ActorYear,index=['Actor','Year'])
ActorYear = ActorYear.T.drop_duplicates()#移除重複項
gp = ActorYear.groupby(by='Actor')
ActorYear['MaxGapYear'] = gp['Year'].transform(lambda x:x.max()-x.min())#添加MaxGapYear在原dataframe後面
ActorYear = ActorYear.drop("Year",axis=1).drop_duplicates()
print("Top-3 directors who collaborate with the most actors",end="?\n")
ActorYear.nlargest(columns='MaxGapYear',n=1,keep='all')
```
