## Swiss Banknote Dataset

Six variables measured on 100 genuine and 100 counterfeit old Swiss 1000-franc
bank notes. The data stem from Flury and Riedwyl (1988). The columns correspond
to the following one label and six variables:

0. Conterfeit - 0: genuine bank notes、1: genuine bank notes (Labels)


1. Length - Length of the bank note


2. Left - Height of the bank note, measured on the left


3. Right - Height of the bank note, measured on the right


4. Bottom - Distance of inner frame to the lower border


5. Top - Distance of inner frame to the upper border


6. Diagonal - Length of the diagonal



More details: [Kaggle | Swiss banknote conterfeit detection](https://www.kaggle.com/chrizzles/swiss-banknote-conterfeit-detection)

```python
# Use [Pandas] to import data | banknotes.csv
# Put your code here 

bank = pd.DataFrame(pd.read_csv('banknotes.csv'))
label = bank.conterfeit

bank.head()
```

```python
# Use [Numpy function] to stack the following columns
# The 5-th column "Bottom" as X, the 6-th column "Top" as Y
# Put your code here

data_xy = np.array(bank[['Bottom','Top']].values)#這邊似乎用bank[['Bottom','Top']].values就可以了但題目說要用numpy
#data_xy[0,:] 測試用
```

```python
# Use [Numpy function] to find where label = 0 / label = 1
# Put your code here

index_0 = np.where(label==0)
index_1 = np.where(label==1)


# Use data_xy to plot the scatter plot 
# Label = 0 → color = green | Label = 1 → color = red
# Fill X and Y in the plt.scatter 
# Remember to put on grid and legend and equal axis
# Put your code here
plt.figure(figsize = (10, 10))
plt.scatter(
                x=data_xy[index_0,0],
                y=data_xy[index_0,1],
                label = 'real banknote',
                color = 'green',s=20, alpha=0.8
            )
plt.scatter(
            x=data_xy[index_1,0],
            y=data_xy[index_1,1],
            label = 'fake banknote',
            color = 'red',s=20, alpha=0.8
            )
plt.legend()
plt.grid(color='k')
plt.axis('equal')
        
```

```python
# Use [Numpy function] to calculate the squared distance matrix between each points
# Put your code here

dist_sq = []
for i in range(0,len(data_xy[:,0])):
    for j in range(0,len(data_xy[:,0])):
        dist_sq.append(round(np.sqrt(np.sum(np.square(data_xy[i]-data_xy[j]))),2))#計算歐式距離(有開根號) 範例好像沒開根號
dist_sq = np.array(dist_sq).reshape(len(data_xy[:,0]),len(data_xy[:,0]))#轉array和轉形狀
```

```python
dist_sq
```

```python
def nearest_partition(dist_sq, K):
    
    # Use [Numpy function] to return the "K Nearest Neighbor"
    # Put your code here
    minpoints = []
    #以下為使用pandas
    #dist_DF = pd.DataFrame(dist_sq)轉成DataFrame形式才能使用nsmallest函數 比較方便
    #for i in range(0,len(data_xy[:,0])):
       # minpoints.append(dist_DF[i].nsmallest(K+1,keep='last').index.values)#取前K+1小的距離的點的index(要K+1因為這些距離前幾近的點
                                                                            #有包含自己，但knn的K值不包含自己本身)
    #以下為使用numpy
    for i in range(0,len(data_xy[:,0])):
        minpoints.append(np.argsort(dist_sq[i])[:K+1])#取前K+1小的距離的點的index(要K+1因為這些距離前幾近的點
                                                                            #有包含自己，但knn的K值不包含自己本身)
    return minpoints
#K=3 測試用
#nearest_partition(dist_sq,K) 測試用
```

```python
# Use [Numpy function] to find where label = 0 / label = 1
# Put your code here

index_0 = np.where(label==0)
index_1 = np.where(label==1)

# Use data_xy to plot the scatter plot 
# Label = 0 → color = green | Label = 1 → color = red
# Fill X and Y in the plt.scatter
# Remember to put on grid and legend and equal axis
# Put your code here


plt.figure(figsize = (10, 10))
plt.scatter(
                x=data_xy[index_0,0],
                y=data_xy[index_0,1],
                label = 'real banknote',
                color = 'green',s=20, alpha=0.8
            )
plt.scatter(
            x=data_xy[index_1,0],
            y=data_xy[index_1,1],
            label = 'fake banknote',
            color = 'red',s=20, alpha=0.8
            )
        
# Draw lines from each point to its three nearest neighbors (set K=3)
# Use some zip magic to make it happen (Hint is at below cell)
# You might need loops to generate the plot below
# set parameter [color='black', linewidth=1, alpha=0.5] when you draw lines 
# Put your code here

K = 3
_kneighbor = nearest_partition(dist_sq, K)
for i in range(len(_kneighbor)):#從每個點出發
    for j in range(len(_kneighbor[i])):#該點的每個鄰居
        plt.plot(*zip(data_xy[i], data_xy[_kneighbor[i][j]]), color = 'black',lw = 1,alpha=0.5)

plt.legend()
plt.grid(color='gray',alpha = 0.4)
plt.axis('equal')
```

```python
# Set seed = 10 | Random select three dots from data_xy
# Save dots index in varible "target" 
# Don't need to change below three lines

random.seed(10)
target = np.array([random.randint(0, len(data_xy)) for i in range(3)])
text = ['A', 'B', 'C']


# Use [Numpy function] to find where label = 0 / label = 1
# Put your code here

index_0 = np.where(label==0)
index_1 = np.where(label==1)


# Use data_xy to plot the scatter plot 
# Label = 0 → color = green | Label = 1 → color = red | Target → color = blue
# Fill X and Y in the plt.scatter
# Remember to put on grid and legend and equal axis
# Put your code here

plt.figure(figsize = (12, 12))
plt.scatter(
                x=data_xy[index_0,0],
                y=data_xy[index_0,1],
                label = 'real banknote',
                color = 'green',s=20, alpha=0.8
            )
plt.scatter(
            x=data_xy[index_1,0],
            y=data_xy[index_1,1],
            label = 'fake banknote',
            color = 'red',s=20, alpha=0.8
            )
plt.scatter(
            x=data_xy[target,0],
            y=data_xy[target,1],
            label = 'target dots',
            color = 'blue',s=60) # Target dots

plt.legend()
plt.grid(color='k',alpha=0.4)
plt.axis('equal')

# Draw lines from the selected three point to its three nearest neighbors (set K=3)
# Use some zip magic to make it happen (Hint is at below cell)
# You might need loops to generate the plot below
# set parameter [color='black', linewidth=1, alpha=0.5] when you draw lines 
# Put text A, B, C on each dots (location: Right + 0.1、 Up + 0.1)
# Put your code here


K = 3
_kneighbor = nearest_partition(dist_sq, K)
for i in range(len(target)):#從選出的三個點出發
    for j in range(len(_kneighbor[i])):#該點的每個鄰居
        plt.plot(*zip(data_xy[target[i]], data_xy[_kneighbor[target[i]][j]]), color = 'black',lw = 1,alpha=0.5)
        plt.text(data_xy[target[i]][0]+0.1,data_xy[target[i]][1]+0.1,text[i],size = 15)
```

### Question：由上圖進行判斷，回答 ABC 三點各自被 KNN 分為哪一群，並說明原因。

### Your Answer：A點和C點是紅色群(fake banknote);B點則是綠色群(real banknote),原因是他們應該被歸類在他們的knearest neighbor中比較多數量的那一群，顯然A和C就是紅色;而B是綠色

```python
# Hint: How to drow a line between two dots

dots = np.array([[1, 2], [6, 3], [5, 6]])

plt.figure(figsize=(15, 5))

plt.subplot(121)
plt.title('Wrong way')        
for i in range(len(dots)):
    plt.scatter(dots[i][0], dots[i][1], color='red')

for i in range(len(dots)):
    for j in range(len(dots)):
        plt.plot(dots[i], dots[j], color = 'black')

plt.subplot(122)
plt.title('Right way')

for i in range(len(dots)):
    plt.scatter(dots[i][0], dots[i][1], color='red')

for i in range(len(dots)):
    for j in range(len(dots)):
        plt.plot(*zip(dots[i], dots[j]), color = 'blue')
```
