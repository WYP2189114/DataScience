# crawler THSR information
Let's start to crawl the imformation of each station of THSR!
* Target web: [THSR Homepage](https://www.thsrc.com.tw)  
* Choose each station:
    1. Homepage -> Travel Information -> Stations
    2. choose different stations by changing url
* Task:
    1. crawler the name, address, operation hours and ticketing hours of each station
    2. organize into a DataFrame

```python
# setting each station name and their url
url = "https://www.thsrc.com.tw/ArticleContent/2f940836-cedc-41ef-8e28-c2336ac8fe68"
StationName = []
StationURL = []
r = requests.get(url)
sp = BeautifulSoup(r.text,'html.parser')
data = sp.find('div',id='nav-tab')
name = data.select('a')
for i in range(0,len(name)):
    StationName.append(name[i].get('title'))
#     print(name[i].get('title'))
    StationURL.append("https://www.thsrc.com.tw"+name[i].get('href'))
#     print(name[i].get('href'))
StationName = pd.Series(StationName)
StationURL = pd.Series(StationURL)
# print(StationName.values[0])
# print(StationURL)
Station = pd.concat([StationName,StationURL],axis=1)
Station.columns=['station','網址']
Station.values
```

```python
# crawler the address of each station
address = []
for i in range(0,Station.shape[0]):
    r = requests.get(Station['網址'].iloc[i])
    sp = BeautifulSoup(r.text,'html.parser')
    data = sp.find('table',summary='車站基本資訊，包含地址、營業時間與售票時間')
#     print(data)
    info = data.select('td')
    address.append(info[0].text.strip('\n'))
#     print(info[0].text)
address = pd.Series(address)
```

```python
# crawler the operation hours and ticketing hours of each station
OpHours = []
TiHours = []
for i in range(0,Station.shape[0]):
    r = requests.get(Station['網址'].iloc[i])
    sp = BeautifulSoup(r.text,'html.parser')
    data = sp.find('table',summary='車站基本資訊，包含地址、營業時間與售票時間')
#     print(data)
    info = data.select('td')
    OpHours.append(info[1].text[5:])
    TiHours.append(info[2].text[5:])
# print(OpHours)
# print(TiHours)
OpHours = pd.Series(OpHours)
TiHours = pd.Series(TiHours)
```

```python
# create a DataFrame
Allinfo = pd.concat([StationName,address,OpHours,TiHours],axis=1)
Allinfo.columns = ["station","address","operation_hours","ticketing_hours"]
Allinfo
```

# Crawler GDP and CPI
Let's start to crawl GDP and CPI!
* Target web: 
    1. [Wiki GDP](https://en.wikipedia.org/wiki/List_of_countries_by_GDP_(nominal))  
    2. [Wiki CPI](https://en.wikipedia.org/wiki/Corruption_Perceptions_Index)
* Task:
    1. crawler GDP table (top 50: United States ~ New Zealand)
    2. organize into a DataFrame (columns: Country, Region, IMF_Estimate, IMF_Year, United_Nations_Estimate, United_Nations_Year, World_Bank_Estimate, World_Bank_Year)
    3. crawler CPI table which contains country and 2020 CPI (top 100: Denmark ~ Suriname)
    4. organize into a DataFrame (columns: Country, CPI_2020)
    5. merge GDP(DataFrame) and CPI(DataFrame), based on Country of GDP
    6. plot and text the names of GDP top 10 countrys

```python
# crawler GDP
urlGDP = "https://en.wikipedia.org/wiki/List_of_countries_by_GDP_(nominal)"
urlCPI = "https://en.wikipedia.org/wiki/Corruption_Perceptions_Index"
r = requests.get(urlGDP)
sp = BeautifulSoup(r.text,'html.parser')
data = sp.find('table',class_="wikitable sortable static-row-numbers plainrowheaders srn-white-background").select("tbody tr")
country = []
Region = []
IMF_Estimate = []
IMF_Year = []
United_Nations_Estimate = []
United_Nations_Year=[]
World_Bank_Estimate=[]
World_Bank_Year=[]
for i in range(2,52):
    country.append(data[i].select('td')[0].text.strip("\xa0"))
    Region.append(data[i].select('td')[1].text.strip("\xa0"))
    IMF_Estimate.append(data[i].select('td')[2].text.strip("\xa0"))
    IMF_Year.append(data[i].select('td')[3].text.strip("\xa0")[-4:])
    United_Nations_Estimate.append(data[i].select('td')[4].text.strip("\xa0"))
    United_Nations_Year.append(data[i].select('td')[5].text.strip("\xa0")[-4:])
    World_Bank_Estimate.append(data[i].select('td')[6].text.strip("\xa0").replace(',',''))
    World_Bank_Year.append(data[i].select('td')[7].text.strip("\n"))
#     data[i].select('td')[0].text
#     data[i].select('td')[1].text
#     data[i].select('td')[2].text
#     data[i].select('td')[3].text
#     data[i].select('td')[4].text
#     data[i].select('td')[5].text
#     data[i].select('td')[6].text
#     data[i].select('td')[7].text
country = pd.Series(country)
Region = pd.Series(Region)
IMF_Estimate = pd.Series(IMF_Estimate)
IMF_Year = pd.Series(IMF_Year)
United_Nations_Estimate = pd.Series(United_Nations_Estimate)
United_Nations_Year= pd.Series(United_Nations_Year)
World_Bank_Estimate= pd.Series(World_Bank_Estimate)
World_Bank_Year= pd.Series(World_Bank_Year)   
```

```python
# create GDP DataFrame
GDPdf = pd.concat([country,Region,IMF_Estimate,IMF_Year,United_Nations_Estimate,United_Nations_Year,World_Bank_Estimate,World_Bank_Year],axis=1)
GDPdf.columns=["Country","Region","IMF_Estimate","IMF_Year","United_Nations_Estimate","United_Nations_Year","World_Bank_Estimate","World_Bank_Year"]
GDPdf
```

```python
# crawler CPI
country = []
CPI_2020 = []
r = requests.get(urlCPI)
sp = BeautifulSoup(r.text,'html.parser')
data = sp.find_all('table')[4].select("tbody tr")
for i in range(2,102):
    country.append(data[i].select('td')[1].text.strip("\xa0"))
    CPI_2020.append(data[i].select('td')[2].text)
country = pd.Series(country)
CPI_2020 = pd.Series(CPI_2020)
```

```python
# CPI DataFrame
CPIdf = pd.concat([country,CPI_2020],axis=1)
CPIdf.columns = ["Country","CPI_2020"]
CPIdf
```

```python
# merge GDP and CPI DataFrame
df_Total = GDPdf.merge(CPIdf,on=['Country'], how='inner')
df_Total.dropna(subset = ["CPI_2020"],inplace=True)
df_Total
```

```python
# plot
df_Total['CPI_2020'] = df_Total['CPI_2020'].astype('int')
df_Total['World_Bank_Estimate'] = df_Total['World_Bank_Estimate'].astype('float')
df_Total['World_Bank_Estimate'] = df_Total['World_Bank_Estimate']
plt.figure(figsize=(8,5))
sns.scatterplot(x=df_Total.CPI_2020, y=df_Total.World_Bank_Estimate,hue=df_Total.Region)
for i in range(0,10):
    plt.text(df_Total.CPI_2020[i]+0.5,df_Total.World_Bank_Estimate[i]+0.1,df_Total.Country[i],size=10)
```
